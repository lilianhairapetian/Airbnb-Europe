# Clustering

```python
#Steps:
# getting only numeric values from data set for clustering data
# creating the cluster 
# then adding the cluster column back to the main data where location Country and city is also available
```

```python
df_nums = city_rec [['Price','Cleanliness Rating',
                  'Guest Satisfaction','City Center (km)',
                  'Metro Distance (km)']]
```

```python
num_cl = range(1, 10) # Defines the range of potential clusters in the data.
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
score = [kmeans[i].fit(df_nums).score(df_nums) for i in range(len(kmeans))] # Creates a score that represents 
# a rate of variation for the given cluster option.

score
```
