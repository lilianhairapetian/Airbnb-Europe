## Categorical Plot

```python
airbnb['Price'].describe()
```

```python
# create a histogram to visualize the distribution of variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution
sns.histplot(airbnb['Price'], bins = 100, kde = True)
```

```python
sub_numeric_smaller.loc[(sub_numeric_smaller['Price'] >= 30) & (sub_numeric_smaller['Price'] <= 80), 'Price category'] = 'Low'
```

```python
sub_numeric_smaller.loc[(sub_numeric_smaller['Price'] >= 80) & (sub_numeric_smaller['Price'] <= 200), 'Price category'] = 'Medium-Low'
```

```python
sub_numeric_smaller.loc[(sub_numeric_smaller['Price'] >= 200) & (sub_numeric_smaller['Price'] <= 500), 'Price category'] = 'Medium'
```

```python
sub_numeric_smaller.loc[(sub_numeric_smaller['Price'] >= 500) & (sub_numeric_smaller['Price'] <= 1000), 'Price category'] = 'Medium-high'
```

```python
sub_numeric_smaller.loc[(sub_numeric_smaller['Price'] >= 1000) & (sub_numeric_smaller['Price'] <= 3000), 'Price category'] = 'High'
```

```python
sub_numeric_smaller.loc[sub_numeric_smaller['Price'] >=3001, 'Price category'] = 'Luxury/High'
```

```python
sub_numeric_smaller['Price category'].value_counts(dropna = False)
```

```python
sub_numeric_smaller.columns
```

### Creating a categorical plot

```python
plot_1 = sns.catplot(x="Cleanliness Rating", y="Price", hue="Price category", data=sub_numeric_smaller)
plt.show()
```

```python
plot_1.savefig("cleanliness&price.png") 
```

#### on categorical plot of cleanliness and price above, it's shown that price category gets more 'high' labels as the cleanliness rating hits the 8.0. Also, majority of luxury accomodations have the highest cleanliness rating (10.0).
